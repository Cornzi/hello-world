```python
from fastai.vision.all import *
from fastai.vision.widgets import *

btn_upload=widgets.FileUpload()
btn_run=widgets.Button(description="Classify!")
btn_rem=widgets.Button(description="Remove")
out=widgets.Output()
model=load_learner("export.pkl")
label=widgets.Label("Prediction will show here!")
def classify(change):
  out.clear_output()
  try:
    img=PILImage.create(btn_upload.data[0])
  except:
    label.value="Sorry! Unsupported Image type!"
    time.sleep(4)
    label.value="Prediction will show here!"
    return
  with out:
    display(img.to_thumb(128))
  pred,te,prob=model.predict(img)
  label.value=f"Prediction:{pred}; Probability:{prob.max().item()}"
def remove(change):
  out.clear_output();
  btn_upload.data=[]
btn_run.on_click(classify)
btn_rem.on_click(remove)
#text=widgets.Text("Enjoy Classifying digits!! Click the button below!")
VBox([widgets.Label("Choose your image"),out,btn_upload,btn_run,btn_rem,label])
```
